```python
import tensorflow as tf
import numpy as np
input_layer_1_np = np.array([5, 10, 15], dtype=np.float32)
input_layer_2_np = np.array([2, 4, 6], dtype=np.float32)
weights_layer_1_np = np.array([0.1, 0.2, 0.3], dtype=np.float32)
weights_layer_2_np = np.array([0.4, 0.1, 0.2], dtype=np.float32)
output_1_np = np.dot(input_layer_1_np, weights_layer_1_np)
output_2_np = np.dot(input_layer_2_np, weights_layer_2_np)
final_output_np = output_1_np + output_2_np
final_output_sigmoid_np = 1 / (1 + np.exp(-final_output_np))

print(f"Numpy output_1: {output_1_np}")
print(f"Numpy output_2: {output_2_np}")
print(f"Numpy final_output (before sigmoid): {final_output_np}")
print(f"Numpy final_output (after sigmoid): {final_output_sigmoid_np}\n")
```

```python

input1 = tf.keras.Input(shape=(3,), name='Input_Layer_1')
input2 = tf.keras.Input(shape=(3,), name='Input_Layer_2')
output_branch1 = tf.keras.layers.Dense(
    1,
    use_bias=False,
    kernel_initializer=tf.constant_initializer(weights_layer_1_np.reshape(-1, 1)),
    trainable=False,
    name='Dot_Product_1'
)(input1)
output_branch2 = tf.keras.layers.Dense(
    1,
    use_bias=False,
    kernel_initializer=tf.constant_initializer(weights_layer_2_np.reshape(-1, 1)),
    trainable=False,
    name='Dot_Product_2'
)(input2)
merged_output = tf.keras.layers.Add(name='Sum_Outputs')([output_branch1, output_branch2])
final_output_activated = tf.keras.layers.Activation('sigmoid', name='Sigmoid_Activation')(merged_output)
model = tf.keras.Model(inputs=[input1, input2], outputs=final_output_activated)
print("Model Summary:")
model.summary()
```

```python

tf_input_1 = tf.constant(input_layer_1_np[np.newaxis, :], dtype=tf.float32)
tf_input_2 = tf.constant(input_layer_2_np[np.newaxis, :], dtype=tf.float32)
tf_prediction_sigmoid = model([tf_input_1, tf_input_2])

print(f"\nTensorFlow Model Output (after sigmoid): {tf_prediction_sigmoid.numpy().flatten()[0]}")
model_intermediate = tf.keras.Model(inputs=[input1, input2], outputs=merged_output)
tf_prediction_raw = model_intermediate([tf_input_1, tf_input_2])
print(f"TensorFlow Model Output (before sigmoid): {tf_prediction_raw.numpy().flatten()[0]}")
```
