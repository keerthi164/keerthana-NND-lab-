```python
import torch

my_array = [10, 20, 30]
bias_array = [0.1, 0.5, -0.2]
my_array_torch = torch.tensor([10., 20., 30.])
print(f"my_array_torch: {my_array_torch}")

my_new_array_torch = torch.tensor([1., 2., 3., 4., 5., 6.])
print(f"my_new_array_torch: {my_new_array_torch}")

bias_array_torch = torch.tensor([0.1, 0.5, -0.2])
print(f"bias_array_torch: {bias_array_torch}")

x_torch = torch.tensor(my_array, dtype=torch.float32)
y_torch = torch.tensor([0.5, 0.2, 0.3])
b_torch = torch.tensor(bias_array)
z_torch = torch.dot(x_torch, y_torch) + b_torch
print(f"Original z (PyTorch): {z_torch}")
def sigmoid_torch(x):
    return 1 / (1 + torch.exp(-x))

sigmoid_z_torch = sigmoid_torch(z_torch)
print(f"Sigmoid of z (PyTorch): {sigmoid_z_torch}")
```

```python
import torch

input_layer_1_torch = torch.tensor([5., 10., 15.])
input_layer_2_torch = torch.tensor([2., 4., 6.])
weights_layer_1_torch = torch.tensor([0.1, 0.2, 0.3])
weights_layer_2_torch = torch.tensor([0.4, 0.1, 0.2])
output_1_torch = torch.dot(input_layer_1_torch, weights_layer_1_torch)
output_2_torch = torch.dot(input_layer_2_torch, weights_layer_2_torch)
final_output_torch = output_1_torch + output_2_torch

print(f"Input Layer 1 (PyTorch): {input_layer_1_torch}")
print(f"Weights Layer 1 (PyTorch): {weights_layer_1_torch}")
print(f"Output from Layer 1 (PyTorch): {output_1_torch:.4f}\n")

print(f"Input Layer 2 (PyTorch): {input_layer_2_torch}")
print(f"Weights Layer 2 (PyTorch): {weights_layer_2_torch}")
print(f"Output from Layer 2 (PyTorch): {output_2_torch:.4f}\n")

print(f"Final Output (PyTorch): {final_output_torch:.4f}")
```
